```python
#removing all the records where city is NaN since its a string
df=df.dropna(subset=['City'])
```

```python
df.isna().sum()
```

```python
df["Items_Count"].skew()
```

```python
#Items_Count showed near-zero skewness (-0.002), indicating symmetric distribution.
#Missing values were imputed using median to maintain robustness.
df["Items_Count"]=df["Items_Count"].fillna(df["Items_Count"].median())
```

```python
df.isna().sum()
```

```python
df["Customer_Rating"].skew()
```

```python
#Customer_Rating shows near-zero skewness (-0.048), indicating a fairly symmetric distribution.
#Missing values were imputed using the median,
#as it is more appropriate for ordinal rating data and robust to extremes.
df["Customer_Rating"]=df["Customer_Rating"].fillna(df["Customer_Rating"].median())
```

```python
df["Delivery_Partner_Rating"].skew()
```

```python
#Delivery_Partner_Rating shows no significant skewness.
#Missing values can be safely imputed using the median,
#as it is suitable for ordinal rating data.
df["Delivery_Partner_Rating"]=df["Delivery_Partner_Rating"].fillna(df["Delivery_Partner_Rating"].median())
```

```python
# EDA ANALYSIS
# basic business KPIs
```

```python
# Total revenue
df['Order_Value'].sum()
```

```python
# Average order value
df["Order_Value"].mean()
```

```python
# Total Orders
df["Order_ID"].nunique()
```

```python
# City-wise Performance
# Revenue by City
df.groupby("City")["Order_Value"].sum().sort_values(ascending=False)
```

```python
# Avg Delivery Time by City
df.groupby("City")["Delivery_Time_Min"].mean().sort_values()
```

```python
#Avg Customer Rating by City
df.groupby("City")["Customer_Rating"].mean().sort_values(ascending=False)
```

```python
# Discount Impact
# This code calculates the average order value for orders with and without discounts.
# It helps determine whether applying discounts increases overall customer spending.
df.groupby("Discount_Applied")["Order_Value"].mean()
```

```python
# This code computes the average number of items purchased for discounted and non-discounted orders.
# It helps assess whether discounts encourage customers to buy more items per order.
df.groupby("Discount_Applied")["Items_Count"].mean()
```

```python
# Correlation Check
# This code checks how delivery time and distance are related to customer ratings.
# It helps see if slower or longer deliveries lead to lower ratings.
df[["Delivery_Time_Min","Customer_Rating","Distance_Km"]].corr()
```
